## Заполнение пропусков

```python
# Копирование и создание нового df для средних значений
df_ef_mean = df_ef.copy()

# заполнение NaN в столбцах с 1 по 11 средним значением
df_ef_mean[df_ef_mean.columns[1:11]] = df_ef_mean[df_ef_mean.columns[1:11]].fillna(df_ef_mean[df_ef_mean.columns[1:11]].mean())

display(df_ef_mean.info())
display(df_ef_mean)
```

<h3> Краткий вывод: </h3>

<ul>
  <li>были заполнены пропуски</li>
  <li>display(df_ef_mean) Показал, что пропуски были заполненны одиннаковыми (для столбца, в которо пропуск) данными. Как пример unit_1 и его значение в 0 строке = [39.713383] и его значение в строке 2 = [39.713383]</li>
</ul> 


## Работа с данными (добавление новых столбцов)

```python
# Название новых столбцов
new_columns = [
    'min',
    'median',
    'max',
]
display(df_ef_mean)

#for i in range (3):
    #df_ef_mean[new_columns[i]] = df_ef_mean.iloc[:, 1:11].min(axis=1)
    #размечтался

# Добавление новых столбцов
df_ef_mean[new_columns[0]] = df_ef_mean.iloc[:, 1:11].min(axis=1)
df_ef_mean[new_columns[1]] = df_ef_mean.iloc[:, 1:11].median(axis=1)
df_ef_mean[new_columns[2]] = df_ef_mean.iloc[:, 1:11].max(axis=1)
df_ef_mean['mean/std'] = df_ef_mean.iloc[:,1:11].mean(axis=1).round(2).astype(str) + '/' + df_ef_mean.iloc[:,1:11].std(axis=1).round(2).astype(str)

# Добавление строки с данными о среднем значении по столбцам
average_annual = df_ef_mean[df_ef_mean.columns[1:-1]].mean()
average_annual['date'] = 'average_annual'
df_ef_mean.loc[len(df_ef_mean)] = average_annual


display(df_ef_mean)
```
